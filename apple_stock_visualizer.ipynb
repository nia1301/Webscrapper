```python
price = alt.Chart(data.reset_index()).mark_line().encode(
    x='Date:T',
    y='Close',
    tooltip=['Date' , 'Open', 'Close','High', 'Low' , 'Volume']).properties(
    width=600,height=200)
price
```

```python
price = alt.Chart(data.reset_index()).mark_line().encode(
    x='Date:T',
    y='Close',
    tooltip=['Date' , 'Open', 'Close','High', 'Low' , 'Volume']).properties(
    width=600,height=200)

volume = alt.Chart(data.reset_index()).mark_line().encode(
    x='Date:T',
    y='Volume'
).properties(
    width=600,height=60)

price & volume
```

```python
price = alt.Chart(data.reset_index()).mark_line().encode(
    x='yearmonth(Date):T',
    y='mean(Close)',
    tooltip=['Date' , 'Open', 'Close','High', 'Low' , 'Volume']).properties(
    width=600,height=200)

volume = alt.Chart(data.reset_index()).mark_line().encode(
    x='yearmonth(Date):T',
    y='mean(Volume)'
).properties(
    width=600,height=60)

price & volume
```
